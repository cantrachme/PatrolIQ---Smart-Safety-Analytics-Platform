```python
#                                                        Feature Engineering
```

```python
sample_df['Hour'] = sample_df['Date'].dt.hour
sample_df['Day_of_Week'] = sample_df['Date'].dt.day_name()
sample_df['Month'] = sample_df['Date'].dt.month
```

```python
sample_df['Is_Weekend'] = sample_df['Day_of_Week'].isin(['Saturday', 'Sunday']).astype(int)
sample_df['Is_Weekend'].head()
```

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

sample_df['Season'] = sample_df['Month'].apply(get_season)
sample_df['Season'].head()
```

```python
# Latitude & Longitude
sample_df = sample_df.dropna(subset=['Latitude', 'Longitude'])
```

```python
 # Normalize Coordinates
from sklearn.preprocessing import StandardScaler

geo_scaler = StandardScaler()
sample_df[['Latitude_scaled', 'Longitude_scaled']] = geo_scaler.fit_transform(
    sample_df[['Latitude', 'Longitude']]
) # Without scaling, longitude dominates distance calculations and breaks clustering.
```

```python
# District / Ward Encoding (Administrative Geography)
sample_df['District'] = sample_df['District'].astype('Int64')
sample_df['Ward'] = sample_df['Ward'].astype('Int64')
```

```python
# Severity Mapping
severity_map = {
    'HOMICIDE': 10,
    'CRIM SEXUAL ASSAULT': 9,
    'ROBBERY': 8,
    'AGGRAVATED ASSAULT': 8,
    'ASSAULT': 7,
    'BATTERY': 7,
    'BURGLARY': 6,
    'MOTOR VEHICLE THEFT': 6,
    'THEFT': 5,
    'NARCOTICS': 5,
    'CRIMINAL DAMAGE': 4,
    'WEAPONS VIOLATION': 7,
    'PUBLIC PEACE VIOLATION': 3,
    'DECEPTIVE PRACTICE': 4,
    'CRIMINAL TRESPASS': 3
}

sample_df['Crime_Severity_Score'] = sample_df['Primary Type'].map(severity_map)
sample_df['Crime_Severity_Score'] = sample_df['Crime_Severity_Score'].fillna(3)
```

```python
# Categorical Encoding


# Encode Crime Type
from sklearn.preprocessing import LabelEncoder

le_crime = LabelEncoder()
sample_df['Primary_Type_Encoded'] = le_crime.fit_transform(sample_df['Primary Type'])



# Encode Location Description
le_location = LabelEncoder()
sample_df['Location_Description_Encoded'] = le_location.fit_transform(
    sample_df['Location Description'].astype(str)
)
```

```python
# Normalize Numerical Features
from sklearn.preprocessing import StandardScaler

num_features = [
    'Hour',
    'Month',
    'Crime_Severity_Score',
    'Latitude',
    'Longitude'
]

scaler = StandardScaler()
sample_df[num_features] = scaler.fit_transform(sample_df[num_features])
```
