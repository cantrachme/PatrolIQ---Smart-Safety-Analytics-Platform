                                                          Feature Engineering 

Encoding Day of Week

```python
day_mapping = {
    'Monday': 0, 'Tuesday': 1, 'Wednesday': 2,
    'Thursday': 3, 'Friday': 4, 'Saturday': 5, 'Sunday': 6
}

df['Day_of_Week_Num'] = df['Day_of_Week'].map(day_mapping)
df['Day_of_Week_Num'].head()
```

Encode Season

```python
def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

df['Season'] = df['Month'].apply(get_season)
```

```python
season_mapping = {
    'Winter': 0,
    'Spring': 1,
    'Summer': 2,
    'Fall': 3
}

df['Season_Num'] = df['Season'].map(season_mapping)
```

```python
df[['Month', 'Season', 'Season_Num']].head(10)
```

Convert Boolean to Numeric

```python
df['Is_Weekend'] = df['Is_Weekend'].astype(int)
df['Arrest'] = df['Arrest'].astype(int)
df['Domestic'] = df['Domestic'].astype(int)
```

Geographic Feature Engineering
- Scale Latitude & Longitude

```python
from sklearn.preprocessing import StandardScaler

geo_scaler = StandardScaler()
df[['Lat_Scaled', 'Long_Scaled']] = geo_scaler.fit_transform(
    df[['Latitude', 'Longitude']]
)
```

- Encode District & Ward

                                                        Clustering

K Means Clustering

- Features 

```python
geo_features = df[['Lat_Scaled', 'Long_Scaled']]
```

- Choose Optimal K (Elbow Method)

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(geo_features)
    inertia.append(kmeans.inertia_)

plt.plot(K_range, inertia, marker='o')
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal K")
plt.show()
```

- Apply K-Means

```python
kmeans = KMeans(n_clusters=6, random_state=42)
df['KMeans_Cluster'] = kmeans.fit_predict(geo_features)
```

- Evaluate K-Means

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score

silhouette_km = silhouette_score(geo_features, df['KMeans_Cluster'])
db_km = davies_bouldin_score(geo_features, df['KMeans_Cluster'])

silhouette_km, db_km
```

                                                  Temporal Pattern Clustering

- Features Used for Temporal Clustering

```python
temporal_features = df[
    ['Hour', 'Day_of_Week_Num', 'Month', 'Is_Weekend']
]
```

- Scale Temporal Features

```python
from sklearn.preprocessing import StandardScaler

temporal_scaler = StandardScaler()
temporal_scaled = temporal_scaler.fit_transform(temporal_features)
```

- Choose Number of Temporal Clusters (Elbow Method)

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
K_range = range(2, 8)

for k in K_range:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(temporal_scaled)
    inertia.append(km.inertia_)

plt.figure(figsize=(6,4))
plt.plot(K_range, inertia, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method for Temporal Clustering")
plt.show()
```

- Apply K-Means for Temporal Clustering

```python
kmeans_time = KMeans(n_clusters=4, random_state=42)
df['Temporal_Cluster'] = kmeans_time.fit_predict(temporal_scaled)
```

- Interpret Temporal Clusters

```python
temporal_profile = df.groupby('Temporal_Cluster')[
    ['Hour', 'Day_of_Week_Num', 'Month', 'Is_Weekend']
].mean()

temporal_profile
```

```python
cluster_labels = {
    0: 'Late Night Crimes',
    1: 'Morning Rush Crimes',
    2: 'Evening Rush Crimes',
    3: 'Weekend / Nightlife Crimes'
}

df['Temporal_Label'] = df['Temporal_Cluster'].map(cluster_labels)
df['Temporal_Label']
```

- Visualize Temporal Clusters (Hour vs Cluster)

```python
plt.figure(figsize=(8,4))
df.boxplot(column='Hour', by='Temporal_Label')
plt.title("Crime Hour Distribution by Temporal Cluster")
plt.suptitle("")
plt.xlabel("Temporal Cluster")
plt.ylabel("Hour of Day")
plt.show()
```

                                                        Dimensionality Reduction

PCA (Principal Component Analysis)

- Select Features for PCA

```python
pca_features = df[
    [
        'Hour',
        'Day_of_Week_Num',
        'Month',
        'Is_Weekend',
        'Lat_Scaled',
        'Long_Scaled',
        'Crime_Severity_Score'
    ]
]
```

- Standardize Features

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
pca_scaled = scaler.fit_transform(pca_features)
```

- Apply PCA (2 Components)

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=42)
pca_components = pca.fit_transform(pca_scaled)
```

- Variance

```python
pca.explained_variance_ratio_
```

- Create PCA DataFrame

```python
pca_df = df.copy()
pca_df['PC1'] = pca_components[:, 0]
pca_df['PC2'] = pca_components[:, 1]
```

- PCA Visualization

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))
plt.scatter(pca_df['PC1'], pca_df['PC2'], alpha=0.3)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA: Crime Data Reduced to 2D")
plt.show()
```

- PCA Feature Importance

```python
import pandas as pd

pca_loadings = pd.DataFrame(
    pca.components_.T,
    columns=['PC1', 'PC2'],
    index=pca_features.columns
)

pca_loadings
```

```python
# =========================================
# FINAL SAVE CELL (VERY IMPORTANT)
# =========================================

from sklearn.preprocessing import StandardScaler
import pandas as pd

# Ensure Date is datetime
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True, errors="coerce")

# -------------------------------
# Temporal Features
# -------------------------------
df["Hour"] = df["Date"].dt.hour
df["Month"] = df["Date"].dt.month
df["Is_Weekend"] = df["Date"].dt.weekday >= 5

# -------------------------------
# Season Encoding
# -------------------------------
def get_season(month):
    if month in [12, 1, 2]:
        return 0  # Winter
    elif month in [3, 4, 5]:
        return 1  # Spring
    elif month in [6, 7, 8]:
        return 2  # Summer
    else:
        return 3  # Fall

df["Season_Num"] = df["Month"].apply(get_season)

# -------------------------------
# Crime Severity Score
# -------------------------------
severity_map = {
    "HOMICIDE": 10,
    "CRIM SEXUAL ASSAULT": 9,
    "ROBBERY": 8,
    "ASSAULT": 7,
    "BATTERY": 7,
    "BURGLARY": 6,
    "MOTOR VEHICLE THEFT": 6,
    "THEFT": 5,
    "CRIMINAL DAMAGE": 4,
    "NARCOTICS": 4
}

df["Crime_Severity_Score"] = (
    df["Primary Type"].map(severity_map).fillna(3)
)

# -------------------------------
# Scale Geographic Features
# -------------------------------
scaler = StandardScaler()
df[["Lat_Scaled", "Long_Scaled"]] = scaler.fit_transform(
    df[["Latitude", "Longitude"]]
)

# -------------------------------
# FINAL SAVE (THIS FILE IS USED BY ML.py)
# -------------------------------
df.to_csv("chicago_crime_sample_500k_clean.csv", index=False)

print("âœ… Notebook data saved successfully")
print("ðŸ“ File created: chicago_crime_sample_500k_clean.csv")
```
